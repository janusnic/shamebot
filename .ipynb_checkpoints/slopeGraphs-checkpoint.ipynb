````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
from matplotlib import rc
rc('text', usetex=False)                   
from matplotlib.lines import Line2D    

def colorscale(hexstr, scalefactor):
    
    def clamp(val, minimum=0, maximum=255):
        if val < minimum:
            return minimum
        if val > maximum:
            return maximum
        return val
    
    hexstr = hexstr.strip('#')

    if scalefactor < 0 or len(hexstr) != 6:
        return hexstr

    r, g, b = int(hexstr[:2], 16), int(hexstr[2:4], 16), int(hexstr[4:], 16)

    r = clamp(r * scalefactor)
    g = clamp(g * scalefactor)
    b = clamp(b * scalefactor)

    return "#%02x%02x%02x" % (r, g, b)

def luminocity(hexstr):
    hexstr = hexstr.strip('#')
    r, g, b = int(hexstr[:2], 16), int(hexstr[2:4], 16), int(hexstr[4:], 16)
    
    return 0.2126*r + 0.7152*g + 0.0722*b

fig, ax = plt.subplots(dpi=300, figsize=(10,10))
ax.axis('equal')

# Width of the "rings" (percentages if the largest "radius"==1)
width = 0.13
kwargs = dict(startangle=90, wedgeprops = {'linewidth': 1})

datas = [[28, 100-28], [45, 100-45], [38, 100-38], [35, 100-35], [35, 100-35], [35, 100-35], [35, 100-35]]
datas = sorted(datas, key=lambda d: d[0]) # Sort so best is in middle

colors = {'DN': '#1aa4d5', 
          'ETC': '#70b94d', 
          'SVT': '#f5c02c', 
          'SVD': '#fc6a21', 
          'Flamman': '#e02718', 
          'Expressen': '#433fa5', 
          'Aftonbladet': '#f66a01'}
sources = colors.keys()
font = "Latin Modern Mono"

props = dict(boxstyle='round', 
             facecolor='white', 
             edgecolor='none',
             alpha=0.0)

rings = []
for i, (data, source) in enumerate(zip(datas, sources)):
    ring, _ = ax.pie(data, 
                     radius=1-width*i, 
                     colors=[colorscale(colors[source], 1.1), colorscale(colors[source], 0.7)], 
                     #labels=data,
                     #labeldistance=0.85,
                     **kwargs)
    
    if luminocity(colors[source]) > 160:
        textcolor = "black"
    else:
        textcolor = "white"
        
        
    textcolor = "white"
    
    ax.text(0.454, 1-width*i*0.40-0.118, u'{}% {}'.format(data[0], source), transform=ax.transAxes, fontsize=14,
        verticalalignment='top',
        horizontalalignment='left', bbox=props, family=font,
        color=textcolor,
        weight="bold")
    
    rings += ring

plt.setp(rings, width=width, edgecolor='white')

#ax.legend(inside[::-1], ['Hardware', 'Software', 'Lolcats'], frameon=False)

props = dict(facecolor='white', 
             edgecolor='none',
             alpha=0.0)

handles = []
for source, data in zip(sources, datas):
    handles.append(mpatches.Patch(color=colors[source], label='{} {}%'.format(source, data[0])))
    
#ax.legend(handles=handles, frameon=False,
#          bbox_to_anchor=(0.85, 0.92), loc=2, borderaxespad=0.)

plt.title('Procent kvinnor\ni svensk dagspress',
         family=font, fontsize=28,
        weight="bold")

plt.xlabel("vecka 32 2015",
          family=font, fontsize=18,
        weight="bold") 
                                      
l = Line2D([0,0],[-1,0], linestyle='--', color="white", linewidth=1)                                    
ax.add_line(l)     

#plt.show()
plt.savefig("cirkle.pdf")
````
